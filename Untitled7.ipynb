```python
# %matplotlib inline

# import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import locale

locale.setlocale(locale.LC_ALL, 'German')

d = "2017-W23"
r = datetime.datetime.strptime(d + '-0', "%Y-W%W-%w")
print(r)

fsk = pd.ExcelFile("Fehlerkennkarte extern NEU.xls")
print(fsk.sheet_names)

dfalle = {sheet_name: fsk.parse(sheet_name) for sheet_name in fsk.sheet_names}

print(dfalle.keys())

df = dfalle[str(71231)]
print(df.head())
df = df.dropna(subset=['Unnamed: 0'])
df.columns = df.iloc[0]
df = df.drop(df.index[0])
df['Datum:'] = pd.to_datetime(df['Datum:'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
df = df.set_index('Datum:').groupby(pd.TimeGrouper('W')).sum()
df.columns.name = None
summe_io = df.columns[20]
summe_nio = df.columns[18]

spalten = [summe_io, summe_nio]

    
print(df[summe_io].loc[r].sum() > 0)
print(df[summe_io].loc[r].sum())
print(df.index)
print("")
print(df[spalten].loc[r])
print("")

colors = ['green', 'red']

ax  = df[spalten].loc[r].plot(kind='bar', rot=0, color=colors)
for p in ax.patches: 
    # ax.annotate(str(p.get_height()), xy=(p.get_x(), p.get_height()))
    ax.annotate(locale.format('%.0f', np.round(p.get_height() ,decimals=0), True), (p.get_x()+p.get_width()/2., p.get_height()), ha='center', va='center', xytext=(0, -10), textcoords='offset points', color='white', fontsize=20, weight='heavy')
print(ax.patches)
plt.tight_layout()
fig = ax.get_figure()

fig.savefig(str(71231)+"-"+d+"-test.pdf")
```
